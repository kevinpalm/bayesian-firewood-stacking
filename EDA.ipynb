```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('firewood.csv')
df.head()
```

```python
print(df.groupby('row')['species'].count())
```

```python
df.boxplot(column='moisture', by='row', figsize=(12, 6), grid=False)
plt.savefig('imgs/moisture_by_row.png')
```

```python
plt.clf()
pt = df.pivot_table(index='row', columns='species', aggfunc='count').fillna(0).astype(int)
pt.columns=pt.columns.droplevel()
del pt.columns.name
fig, axs = plt.subplots(2, 2, figsize=(12, 6))
i = 0
for x in range(1, 3):
    for y in range(1, 3):
        i += 1
        pt.loc[i,:].plot(kind='bar', ax=axs[x-1][y-1], title='Row {}'.format(i), ylim=(0, 13))
plt.tight_layout()
plt.savefig('imgs/species_by_row.png')
```

```python
plt.clf()
df.boxplot(column='moisture', by='species', figsize=(12, 6), grid=False)
plt.savefig('imgs/moisture_by_species.png')
```
